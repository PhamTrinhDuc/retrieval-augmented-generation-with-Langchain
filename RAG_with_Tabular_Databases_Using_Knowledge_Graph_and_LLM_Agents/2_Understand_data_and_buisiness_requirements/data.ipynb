# **2. Physicians.csv**
```
physician_id: id định danh cho bác sĩ (unique)
physician_name: tên bác sĩ.
physician_dob: sinh nhật bác sĩ.
physician_grad_year: năm tốt nghiệp trường y.
medical_school: nơi bác sĩ theo học trường y.
salary: lương của bác sĩ.
```

````python
PHYSICIAN_DATA_PATH = "data/physicians.csv"
data_physician = pd.read_csv(PHYSICIAN_DATA_PATH)

print(data_physician.shape)
data_physician.head()
````

````python
data_physician.info()
````

````python
# Tìm các bác sĩ tốt nghiệp từ Harvard Medical School
data_physician[data_physician['medical_school'] == "Harvard Medical School"]
````

````python
# Tìm bác sĩ có lương cao nhất
highest_salary = data_physician[data_physician['salary'] == data_physician['salary'].max()]
print("\nBác sĩ có lương cao nhất:")
print(highest_salary)
# Tính lương trung bình của các bác sĩ
print("\nLương trung bình của các bác sĩ")
data_physician['salary'].mean()
````

````python
# Đếm số lượng bác sĩ theo trường y khoa

school_counts = data_physician['medical_school'].value_counts()
print("\nSố lượng bác sĩ theo trường y khoa:")
print(school_counts)
````

````python
# Đếm số lượng bác sĩ theo năm tốt nghiệp

grad_year_counts = data_physician['physician_grad_year'].value_counts()
print("\nSố lượng bác sĩ theo năm tốt nghiệp:")
print(grad_year_counts)
````

# **6. Visits.csv**
```
visit_id: The unique identifier of a hospital visit.
patient_id: The ID of the patient associated with the visit.
date_of_admission: The date the patient was admitted to the hospital.
room_number: The patient’s room number.
admission_type: One of ‘Elective’, ‘Emergency’, or ‘Urgent’.
chief_complaint: A string describing the patient’s primary reason for being at the hospital.
primary_diagnosis: A string describing the primary diagnosis made by the physician.
treatment_description: A text summary of the treatment given by the physician.
test_results: One of ‘Inconclusive’, ‘Normal’, or ‘Abnormal’.
discharge_date: The date the patient was discharged from the hospital
physician_id: The ID of the physician that treated the patient.
hospital_id: The ID of the hospital the patient stayed at.
payer_id: The ID of the insurance payer used by the patient.
billing_amount: The amount of money billed to the payer for the visit.
visit_status: One of ‘OPEN’ or ‘DISCHARGED’.
```

````python
data_visit = pd.read_csv("data/visits.csv")
data_visit.head()
````

````python
data_visit.info()
````

````python
# 3 cột null là chief_complaint, treatment_description, primary_diagnosis
data_visit['chief_complaint'].notnull().value_counts()
````

````python
print(data_visit['billing_amount'].max())
print(data_visit['billing_amount'].min())
print(data_visit['billing_amount'].mean())
````
